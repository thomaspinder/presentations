```julia
using Distributions, Gadfly
Gadfly.set_default_plot_size(10inch, 10inch)
ex_theme = Theme(
    line_width=0.75mm,
    panel_fill="white");
```

```julia
p0 = plot(x=rand(Normal(20, 10), 100000), Geom.density(bandwidth=1), Guide.title("Prior for Î²0"), Coord.cartesian(ymin=0, ymax=0.15), ex_theme)
p1 = plot(x=rand(Normal(10, 4), 100000), Geom.density(bandwidth=1), Guide.title("Prior for Î²1"), Coord.cartesian(ymin=0, ymax=0.15), ex_theme)
p2 = plot(x=rand(Normal(0, 50), 100000), Geom.density(bandwidth=1), Guide.title("Prior for Î²2"), Coord.cartesian(ymin=0, ymax=0.15), ex_theme)
pÏƒ = plot(x=rand(InverseGamma(2, 10), 10000), Geom.density(bandwidth=1), Guide.title("Prior for Ïƒ"), Coord.cartesian(ymin=0, ymax=0.15), ex_theme)
gridplot = gridstack([p0 p1; p2 pÏƒ])
```
